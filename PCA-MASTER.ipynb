```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import preprocessing

from sklearn.decomposition import PCA

from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
from sklearn.mixture import GaussianMixture

from sklearn import metrics
```

# Prepare the Data

```
df = pd.read_excel('https://query.data.world/s/kdys6uqv5oazrf3av62phbqyimd2qh')
df.head()
```

```
# (sementara) Hanya menggunakan data numerik
df = df.drop(['Shrt_Desc', 'GmWt_1', 'GmWt_Desc1',	'GmWt_2',	'GmWt_Desc2',	'Refuse_Pct'], axis=1)
df.head()
```

```
df.shape
```

```
df.info()
```

```
df.isnull().sum()
```

```
df.fillna(0, inplace=True)
```

```
products = df['NDB_No'].values
features = df.iloc[:, 1:].values
```

```
features_norm = preprocessing.normalize(features)
```

```
# PCA sebagai representasi fitur
pca_1 = PCA(n_components=0.95)
# pca_1 = PCA()
pca_features = pca_1.fit_transform(features_norm)
pca_features.shape
```

```
# PCA 2 dimensi untuk visualisasi
pca_2 = PCA(n_components = 2)
Z_PCA = pca_2.fit_transform(pca_features)
```

# Clustering I: K-Means

```
silhouette = []

for i in range(2,11):
  kmeans = KMeans(n_clusters=i)
  kmeans.fit(pca_features)
  labels = kmeans.labels_
  silhouette.append(metrics.silhouette_score(pca_features, labels, metric='euclidean'))
```

```
i = list(range(2,11))
plt.figure(figsize=(8, 6))
plt.plot(i, silhouette)
plt.xlabel("clusters")
plt.ylabel("silhouette score")
plt.title("K-Means Performance")
plt.show()
```

```
# Get the indices of maximum silhouette
max_index = np.where(silhouette == np.amax(silhouette))
max_index =  max_index[0] + 2
```

```
# n = max_index[0]
n = 3
kmeans = KMeans(n_clusters=n)
kmeans.fit(pca_features)
labels = kmeans.labels_
silhouette = metrics.silhouette_score(pca_features, labels, metric='euclidean')

plt.figure(figsize=(8, 6))
plt.scatter(Z_PCA[:, 0], Z_PCA[:, 1], c=labels, s=50, cmap='viridis')
plt.title("{} cluster(s) \n Sillouette Score : {:.4f}" .format(n,silhouette))
plt.show()
```

# Clustering II: Agglomerative Hierarchical Clustering

```
# plt.figure(figsize=(15,8))
# plt.ylabel('Distance')
# plt.xlabel('Samples')
# dendrogram = sch.dendrogram(sch.linkage(pca_features, method='ward'))
```

```
silhouette = []

for i in range(2,11):
  agglomerative_hierarchical_clustering_model = AgglomerativeClustering(n_clusters=i)
  clusters = agglomerative_hierarchical_clustering_model.fit_predict(pca_features)
  silhouette.append(metrics.silhouette_score(pca_features, clusters, metric='euclidean'))
```

```
i = list(range(2,11))
plt.figure(figsize=(8, 6))
plt.plot(i, silhouette)
plt.xlabel("clusters")
plt.ylabel("silhouette score")
plt.title("Agglomerative Performance")
plt.show()
```

```
# Get the indices of maximum element in numpy array
max_index = np.where(silhouette == np.amax(silhouette))
max_index =  max_index[0] + 2
```

```
# n = max_index[0]
n = 2
agglomerative_hierarchical_clustering_model = AgglomerativeClustering(n_clusters=n)
clusters = agglomerative_hierarchical_clustering_model.fit_predict(pca_features)
```

```
silhouette = metrics.silhouette_score(pca_features, clusters, metric='euclidean')
```

```
plt.figure(figsize=(8, 6))
plt.scatter(Z_PCA[:, 0], Z_PCA[:, 1], c=clusters, cmap='viridis')
plt.title("{} cluster(s) \n Sillouette Score : {:.4f}" .format(n,silhouette))
plt.show()
```

# Clustering III: Gaussian Mixture Model

```
silhouette = []

for i in range(2,11):
  model = GaussianMixture(n_components=i)
  model.fit(pca_features)
  yhat = model.predict(pca_features)
  silhouette.append(metrics.silhouette_score(pca_features, yhat, metric='euclidean'))
```

```
i = list(range(2,11))
plt.figure(figsize=(8, 6))
plt.plot(i, silhouette)
plt.xlabel("components")
plt.ylabel("silhouette score")
plt.title("GMM performance")
plt.show()
```

```
# Get the indices of maximum element in numpy array
max_index = np.where(silhouette == np.amax(silhouette))
max_index =  max_index[0] + 2
```

```
# n = max_index[0]
n = 3
model = GaussianMixture(n_components=n) # Sesuaikan dengan silhouette tertinggi
model.fit(pca_features)
clusters = model.predict(pca_features)
```

```
silhouette = metrics.silhouette_score(pca_features, clusters, metric='euclidean')
```

```
plt.figure(figsize=(8, 6))
plt.scatter(Z_PCA[:, 0], Z_PCA[:, 1], c=clusters, cmap='viridis')
plt.title("{} cluster(s) \n Sillouette Score : {:.4f}" .format(n,silhouette))
plt.show()
```
